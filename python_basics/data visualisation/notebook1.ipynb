```python
import pandas as pd
#load the csv file
df=pd.read_csv(r"C:\Users\Admin\Downloads\kaggle data\Employe_Performance_dataset.csv")
#display the first 10 rows
print(df.head(10))
```

```python
#display column names
print(df.columns)
#check column datatypes
print(df.dtypes)
#check number of rows and columns
print(df.shape)
```

```python
#Select the Name, Age, and Salary columns using df[]
print(df[['Name','Age','Salary']])
```

```python
#handle missing values and outliers
# Checking for missing values using isnull()
missing_values = df.isnull().sum()

print(missing_values)

# Fill numeric columns with mean
df.fillna(df.mean(numeric_only=True), inplace=True)

# Fill categorical columns with mode
# Fill categorical columns with mode (Corrected Version)
for col in df.select_dtypes(include=["object"]).columns:
    df[col] = df[col].fillna(df[col].mode()[0])

# Confirm missing values are filled
print(df.isnull().sum())
```

```python
#visualizations
import matplotlib.pyplot as plt
import seaborn as sns

#A box plot helps us detect outliers in salary data
plt.figure(figsize=(6, 4))
sns.boxplot(x=df["Salary"], color="lightblue")
plt.title("Salary Box Plot")
plt.xlabel("Salary")
plt.show()


#A histogram that shows the frequency of salary values.
plt.figure(figsize=(8, 5))
df["Salary"].plot(kind="hist", bins=10, color="blue", edgecolor="black")
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()

#A bar chart helps us see how many employees work in each department.
plt.figure(figsize=(8, 5))
df["Department"].value_counts().plot(kind="bar", color="orange")
plt.title("Number of Employees per Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.xticks(rotation=45)  # Rotate labels for better readability
plt.show()

#A pie chart helps us see gender distribution.
plt.figure(figsize=(6, 6))
df["Gender"].value_counts().plot(kind="pie", autopct="%1.1f%%", colors=["lightblue", "pink", "gray"])
plt.title("Gender Distribution")
plt.ylabel("")  # Remove y-axis label
plt.show()

#A scatter plot shows the relationship between salary and years of experience.
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Experience"], y=df["Salary"], color="green")
plt.title("Salary vs Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()

#A heatmap shows how numerical variables are related.
# Select only numeric columns
df_numeric = df.select_dtypes(include=["number"])

# Generate correlation heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()


```
